```python
from pyspark.sql import SparkSession
from pyspark.sql import Row

spark = SparkSession.builder.appName("CreateSampleData").getOrCreate()

data = [
    Row(sale_id=1, store_id=101, sku="SKU001", qty=2, price=999.00, discount=0, sale_ts="2025-11-01T10:12:00", txn_id="TXN0001"),
    Row(sale_id=2, store_id=101, sku="SKU002", qty=1, price=1499.00, discount=100, sale_ts="2025-11-01T10:13:00", txn_id="TXN0002"),
    Row(sale_id=3, store_id=102, sku="SKU001", qty=1, price=999.00, discount=0, sale_ts="2025-11-01T11:20:00", txn_id="TXN0003"),
    Row(sale_id=4, store_id=103, sku="SKU003", qty=3, price=499.00, discount=50, sale_ts="2025-11-01T12:30:00", txn_id="TXN0004"),
]

df = spark.createDataFrame(data)
display(df)
```

```python
raw_path = "/mnt/modastyle/raw/store_sales"

df.write.mode("overwrite").option("header", "true").csv(raw_path)

print("✅ Sample CSV created at:", raw_path)
```
